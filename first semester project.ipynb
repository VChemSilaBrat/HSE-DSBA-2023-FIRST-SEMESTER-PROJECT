```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv("sourse data\salesmonthly.csv", parse_dates=["datum"])
df["Year"] = df["datum"].dt.year # adding a year coloumn
df["Month"] = df['datum'].dt.month # adding a month coloumn
```

```python
# Data describe
# M01AB Sales volume of anti-inflammatory and antirheumatic products, non-steroids, Acetic acid derivatives 
# and related
# M01AE Sales volume of anti-inflammatory and antirheumatic products, non-steroids, Propionic acid derivatives
# N02BA Sales volume of other analgesics and antipyretics, Salicylic acid and derivatives
# N02BE Sales volume of other analgesics and antipyretics, Pyrazolones and Anilides
# N05B Sales volume of psycholeptics drugs, Anxiolytic drugs
# N05C Sales volume of psycholeptics drugs, Hypnotics and sedatives drugs
# R03 Sales volume of drugs for obstructive airway diseases
# R06 Sales volume of antihistamines for systemic use
```

```python
# Data cleanup
df.M01AB = df.M01AB + df.M01AE
df.N02BA = df.N02BA + df.N02BE
df.N05B = df.N05B + df.N05C
df.R03 = df.R03 + df.R06
# grouping by type of drug
df1 = df.rename(columns={"M01AB": "anti_inflammatory", "N02BA": "analgesics", "N05B": "psycholeptics_drugs", "R03": "antihistamines_obstructive_airway"}).drop(["M01AE", "N02BE", "N05C", "R06"], axis='columns')
# drop other types of drugs + rename
df1.info()
```

```python
# Descriptive statistics:
df1.describe(include=[np.number])
```

```python
# for the correct statistic i'll borrow the data from another csv
fd_another = pd.read_csv("sourse data\salesdaily.csv")
fd_another
```

```python
tmp = fd_another[fd_another.Month == 1][fd_another.Year == 2017] # find all days in first month of 2017
tmp
```

```python
tmp.M01AB = tmp.M01AB + tmp.M01AE
tmp.N02BA = tmp.N02BA + tmp.N02BE
tmp.N05B = tmp.N05B + tmp.N05C
tmp.R03 = tmp.R03 + tmp.R06
tmp # make the same summary as for original dataset
```

```python
tmp = ["2017-01-31", sum(tmp.M01AB), sum(tmp.N02BA), sum(tmp.N05B), sum(tmp.R03), 2017, 1]
df1[df1.datum == "2017-01-31"] = tmp
df1[df1.datum == "2017-01-31"]
df1.to_csv("data_parsed.csv", index=False) # saving parsed data for Streamlit
```
